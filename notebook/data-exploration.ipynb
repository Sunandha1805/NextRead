```python
import pandas as pd
import numpy as np
import os
import kagglehub
```

```python
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")
print("Path to dataset files:", path)
print(os.listdir(path))
```

```python
file_path = os.path.join(path, "books.csv")
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['missing_description'] = np.where(df['description'].isna(), 1, 0)
df['age_of_book'] = 2026 - df['published_year']
```

```python
# from ydata_profiling import ProfileReport
# profile = ProfileReport(books, explorative=True)
# profile.to_notebook_iframe()
```

```python
df[(df['description'].isna()) | (df['num_pages'].isna()) | (df['average_rating'].isna()) | (df['published_year'].isna())].shape
```

```python
books = df[~(df['description'].isna()) & ~(df['num_pages'].isna()) & ~(df['average_rating'].isna()) & ~(df['published_year'].isna())]
```

```python
books.head()
```

```python
books.shape
```

```python
books['description_length'] = books['description'].str.split().str.len()
```

```python
books_25_words = books[books['description_length'] >= 25]
```

```python
books_25_words['title_and_subtitle'] = (
    np.where(books_25_words['subtitle'].isna(), books_25_words['title'],
            books_25_words[['title', 'subtitle']].astype(str).agg(" : ".join, axis = 1))
)
```

```python
books_25_words.head()
```

```python
books_25_words['tagged_description'] = books_25_words[['isbn13', 'description']].astype(str).agg(" ".join, axis = 1)
```

```python
books_25_words.head(1)
```

```python
books_25_words.drop(['subtitle', 'missing_description', 'age_of_book', 'description_length'], axis = 1).to_csv("books_cleaned.csv", index=False)
```
